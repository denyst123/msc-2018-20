```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
x=np.linspace(-1,1,2000)
```

```python
y=np.cos(x) + 0.3*np.random.rand(2000)
```

```python
p=np.polynomial.Chebyshev.fit(x,y,90)
```

```python
t=np.linspace(-1,1,200)
```

```python
plt.plot(x, y, 'r.')
```

```python
plt.plot(t,p(t),'k-',lw=4)
```

```python
plt.show()
```
