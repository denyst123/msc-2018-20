Light Curve of The Star 'Algol' (Eclipsing Binary)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Data downloaded from AAVSO

```python
data=pd.read_csv('/home/guest/Desktop/aavsodata_5d5fe30ae47ed.txt',nrows=10000)
data.head()
```

Plotting the light curve of the star,in which 'JD' & 'Magnitude' taken along X & Y axis respectively.

```python
data.plot(x='JD',y='Magnitude',kind='scatter')
plt.gca().invert_yaxis()
```

           Light Curve of Algol

For better analysis,slicing the light curve randomly into 4 parts and comparing

```python
winter = data[['JD','Magnitude']][:150]
autumn = data[['JD','Magnitude']][500:1000]
summer = data[['JD','Magnitude']][2000:5000]
spring = data[['JD','Magnitude']][6500:9500]
```

Sliced Part-1 and Observation

Sliced Part-3 and Observation

```python
summer.plot(y='Magnitude',x='JD',kind='scatter')
```

Sliced Part-4 and Observation

```python
spring.plot(y='Magnitude',x='JD',kind='scatter')
```

Comparing the different sliced parts

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
axes
ax11 = axes[0][0]
ax12 = axes[0][1]
ax21 = axes[1][0]
ax22 = axes[1][1]
line_width = 2.5
winter.plot(x='JD', y='Magnitude',kind='scatter',ax=ax11, c='blue', legend=False, lw=line_width)
autumn.plot(x='JD', y='Magnitude',kind='scatter',ax=ax12, c='red', legend=False, lw=line_width)
summer.plot(x='JD', y='Magnitude',kind='scatter',ax=ax21, c='orange', legend=False, lw=line_width)
spring.plot(x='JD', y='Magnitude',kind='scatter',ax=ax22, c='green', legend=False, lw=line_width)
```
